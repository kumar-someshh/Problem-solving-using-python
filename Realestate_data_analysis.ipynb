**Q1**  How many unknown ("?) values are present in the dataset? Remove/Delete unknown ("?) values present in the dataset to make it null value.

 ## **``learnings``**

* apply()  
Syntax: df.apply(func, axis=0, raw=False, result_type=None, args=(), **kwds)

  Applies a specified function to either rows or columns of a DataFrame.
axis=0 default(for column)


* applymap()

``Usage:`` df.applymap(func) where func is the function to apply to each cell.

``Applications:`` Ideal for element-wise transformations across multiple rows and columns (like scaling, removing outliers, etc.).

``Compared to apply():``

applymap() works on individual cells, while apply() works on whole columns or rows.
applymap() can be slightly faster in some cases due to fewer internal operations.
applymap() is better for simple element-wise operations, while apply() is more versatile for complex column-level tasks.

----------------

* DataFrame.where(cond, other=nan, inplace=False, axis=None, level=None, errors='raise', try_cast=False)

 * cond: The boolean condition to apply to each element.
 - other: The value to replace elements where the condition is False (default is NaN).
 - inplace: If True, modifies the DataFrame in place; otherwise, returns a new DataFrame (default is False).
 - axis: The axis along which to apply the condition (0 for rows, 1 for columns, default is None for both).
 - level: For MultiIndex DataFrames, specifies the level along which to apply the condition.
 - errors: Specifies how to handle errors during casting (e.g., 'raise', 'ignore').
 - try_cast: If True, attempts to cast the result back to the original dtype.

**``learnings``**

reset_index()

 - It resets the index of a DataFrame, either creating a new default integer index or using a specified column as the new index.

 -df = df.reset_index(drop=False, inplace=False, col_level=None, col_fill='')
Parameters:

  - drop=False: (default) Preserves the existing index as a column in the DataFrame.
  - drop=True: Drops the existing index, not retaining it as a column.
  - inplace=False: (default) Returns a new DataFrame with the modified index.
  - inplace=True: Modifies the index of the existing DataFrame directly.
  - col_level=None: Specifies the level of a MultiIndex to use as the new index.
  - col_fill='': Fills missing values in the new index column (if created) with a specified value.
